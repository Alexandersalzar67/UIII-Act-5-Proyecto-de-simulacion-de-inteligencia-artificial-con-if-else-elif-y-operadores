```python
# --- SISTEMA DE IA PARA ANÁLISIS DE YOUTUBERS ---

# 1. Captura de datos (Entrada interactiva)
nombre_canal = input("¿Cuál es el nombre de tu canal de YouTube?: ")
suscriptores = int(input("Ingresa tu cantidad total de suscriptores: "))
vistas_mensuales = int(input("Ingresa el promedio de vistas de tu último mes: "))
tipo_contenido = input("¿Qué contenido creas? (gaming/educativo/vlogs): ")

# 2. Análisis de rango (Estructura if-elif-else con Operadores Relacionales)
# Se utilizan operadores como >= (mayor o igual) para clasificar al creador [4, 8]
if suscriptores >= 1000000:
    rango = "Creador de Élite (Botón de Oro)"
elif suscriptores >= 100000:
    rango = "Creador Consolidado (Botón de Plata)"
elif suscriptores >= 1000:
    rango = "Creador en Crecimiento (Monetizando)"
else:
    rango = "Creador Emergente"

# 3. Evaluación de rendimiento (Operadores Lógicos y Aritméticos)
# Calculamos el ratio de fidelidad (vistas por suscriptor)
ratio_fidelidad = vistas_mensuales / suscriptores if suscriptores > 0 else 0

# La IA decide el estado del canal usando 'and' y 'or' [4, 9]
if vistas_mensuales > 50000 and ratio_fidelidad > 0.5:
    estado_canal = "Canal viral y altamente activo"
elif vistas_mensuales > 10000 or suscriptores > 50000:
    estado_canal = "Canal estable"
else:
    estado_canal = "Canal con necesidad de mayor alcance"

# 4. Recomendación estratégica (Operadores Relacionales de Igualdad)
if tipo_contenido == "gaming":
    estrategia = "Te recomiendo realizar colaboraciones en directos."
elif tipo_contenido == "educativo":
    estrategia = "Deberías crear cursos en plataformas externas."
elif tipo_contenido == "vlogs":
    estrategia = "Enfócate en Shorts para atraer audiencia nueva."
else:
    estrategia = "Sigue analizando las tendencias de tu nicho."

# 5. Salida de resultados (Resumen del análisis)
print("\n" + "="*40)
print(f"ANÁLISIS DE IA PARA: {nombre_canal}")
print(f"Rango detectado: {rango}")
print(f"Estado del canal: {estado_canal}")
print(f"Sugerencia de la IA: {estrategia}")
print("="*40)
```
