```python
# ==========================================================
#   SISTEMA DE INTELIGENCIA ARTIFICIAL PARA YOUTUBERS
# ==========================================================
# Programa interactivo utilizando:
# - if
# - elif
# - else
# - operadores aritméticos
# - operadores relacionales
# - operadores lógicos
#
# RESTRICCIONES:
# ✔ Sin funciones
# ✔ Sin clases
# ✔ Sin ciclos
# ✔ Solo programación secuencial con condicionales
# ==========================================================

print("==============================================")
print("   IA ANALIZADORA DE CANALES DE YOUTUBE")
print("==============================================")

# ---------------- ENTRADA DE DATOS ----------------

nombre = input("Nombre del canal: ")

suscriptores = int(input("Cantidad de suscriptores: "))

vistas = int(input("Promedio de vistas por video: "))

likes = int(input("Cantidad promedio de likes: "))

comentarios = int(input("Cantidad promedio de comentarios: "))

videos = int(input("Cantidad total de videos: "))

# ---------------- PROCESAMIENTO ----------------

engagement = likes + comentarios

promedio = vistas / videos

# ---------------- ANALISIS DE IA ----------------

print("\n==============================================")
print("      RESULTADOS DEL ANALISIS IA")
print("==============================================")

print("Canal analizado:", nombre)
print("Engagement total:", engagement)
print("Promedio de vistas:", promedio)

# -------- ANALISIS DE POPULARIDAD --------

if suscriptores >= 1000000:
    print("Nivel del canal: YOUTUBER FAMOSO")
elif suscriptores >= 100000:
    print("Nivel del canal: YOUTUBER POPULAR")
elif suscriptores >= 10000:
    print("Nivel del canal: YOUTUBER EN CRECIMIENTO")
else:
    print("Nivel del canal: CANAL PEQUEÑO")

# -------- ANALISIS DE RENDIMIENTO --------

if vistas > suscriptores:
    print("La IA detecta videos VIRALIZADOS")
elif vistas == suscriptores:
    print("La IA detecta rendimiento ESTABLE")
else:
    print("La IA detecta bajo alcance")

# -------- ANALISIS DE INTERACCION --------

if engagement >= 50000:
    print("La comunidad es MUY ACTIVA")
elif engagement >= 10000:
    print("La comunidad es ACTIVA")
elif engagement >= 1000:
    print("La comunidad es MODERADA")
else:
    print("La comunidad tiene poca interacción")

# -------- ANALISIS CON OPERADORES LOGICOS --------

if suscriptores >= 100000 and engagement >= 10000:
    print("La IA recomienda activar PATROCINIOS")
else:
    print("La IA recomienda seguir creciendo")

if vistas >= 500000 or likes >= 50000:
    print("La IA detecta ALTO IMPACTO en redes")
else:
    print("La IA detecta impacto normal")

if not(videos >= 100):
    print("El canal todavía tiene pocos videos")
else:
    print("El canal tiene bastante contenido")

# -------- ANALISIS FINAL --------

if suscriptores >= 1000000 and vistas >= 500000 and engagement >= 50000:
    print("\nRESULTADO FINAL:")
    print("La IA considera que el canal es ÉXITO MUNDIAL")

elif suscriptores >= 100000 and vistas >= 100000:
    print("\nRESULTADO FINAL:")
    print("La IA considera que el canal tiene GRAN FUTURO")

elif suscriptores >= 10000:
    print("\nRESULTADO FINAL:")
    print("La IA considera que el canal está CRECIENDO")

else:
    print("\nRESULTADO FINAL:")
    print("La IA recomienda mejorar el contenido")

print("\n==============================================")
print("     FIN DEL SISTEMA DE INTELIGENCIA IA")
print("==============================================")
```
